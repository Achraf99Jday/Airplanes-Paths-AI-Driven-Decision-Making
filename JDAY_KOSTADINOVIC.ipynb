```python
import sys
import heapq


class Noeud:
    def __init__(self, nom):
        self.nom = nom
        self.visite = False
        self.adjacents = []
        self.predecesseurs = None
        self.distance_min = sys.maxsize

    def __lt__(self, autre):
        return self.distance_min < autre.distance_min


class Arc:
    def __init__(self, debut, fin, poids):
        self.debut = debut
        self.fin = fin
        self.poids = poids


# calcule du plus court chemin
def calcul_plus_court_chemin(depart):
    noeuds_a_visiter = []
    depart.distance_min = 0
    heapq.heappush(noeuds_a_visiter, depart)

    while noeuds_a_visiter:
        noeud_actuel = heapq.heappop(noeuds_a_visiter)
        for arc in noeud_actuel.adjacents:
            tmp_dist = arc.debut.distance_min + arc.poids
            if tmp_dist < arc.fin.distance_min:
                arc.fin.distance_min = tmp_dist
                arc.fin.predecesseurs = arc.debut
                heapq.heappush(noeuds_a_visiter, arc.fin)


# renvoie le plus court chemin
def get_plus_court_chemin(depart, arrivee):
    chemin = []
    depart_trouve = False
    while arrivee and not depart_trouve:
        if arrivee.nom == depart.nom:
            depart_trouve = True
        chemin.append(arrivee.nom)
        arrivee = arrivee.predecesseurs
    if not depart_trouve:
        return []
    chemin.reverse()
    return chemin


# donne le plus court chemin
def plus_court_chemin(depart, arrivee):
    # print("Le plus court chemin de {} à {} est : ".format(depart.nom, arrivee.nom))
    calcul_plus_court_chemin(depart)
    chemin = get_plus_court_chemin(depart, arrivee)
    return chemin


# créé les noeuds (classe) selon un graphe en forme de dictionnaire
def creer_noeuds(graphe):
    liste_noeuds = []

    for noeud1 in graphe:
        node1 = Noeud(str(noeud1))
        liste_noeuds.append(node1)

    return liste_noeuds


# vérifie si le node est dans la liste des noeuds
# si oui, renvoie le noeud, sinon renvoie None
def noeud_dans_liste(liste_noeuds, noeud):
    for n in liste_noeuds:
        if str(noeud) == n.nom:
            return n
    return None


# vérifie si le node est dans la liste des noeuds
# si oui, renvoie l'indice du noeud, sinon renvoie None
def indice_noeud_dans_liste(liste_noeuds, noeud):
    for i, n in enumerate(liste_noeuds):
        if str(noeud) == n.nom:
            return i
    return None


# génère un graphe sous forme de classe à partir d'un format dictionnaire
# renvoie la liste des noeuds créés
def generer_graphe(dict_arcs):
    liste_noeuds = creer_noeuds(dict_arcs)

    for noeud1 in dict_arcs:
        node1 = noeud_dans_liste(liste_noeuds, noeud1)
        if node1:
            for noeud2 in dict_arcs[noeud1]:
                node2 = noeud_dans_liste(liste_noeuds, noeud2)
                if node2:
                    edge = Arc(node1, node2, dict_arcs[noeud1][noeud2])
                    node1.adjacents.append(edge)

    return liste_noeuds


# créé les sommets du graphe G~
def sommets_G_transforme(liste_sommets, liste_arcs):
    sommets = []

    for sommet in liste_sommets:
        tmp_jour_in = [0]
        tmp_jour_out = [0]
        for arc in liste_arcs:
            jour = arc[2]
            if sommet == arc[0]:
                if jour not in tmp_jour_in:
                    tmp_jour_in.append(jour)
                    if (sommet, jour) not in sommets:
                        sommets.append((sommet, jour))
            elif sommet == arc[1]:
                if jour not in tmp_jour_out:
                    tmp_jour_out.append(jour)
                    if (sommet, jour + 1) not in sommets:
                        sommets.append((sommet, jour + 1))

    return sommets


# tri les sommets du graphe G~ par sommet
def tri_sommets_G_transforme(liste_sommets, sommets_G):
    sommets_tries = []

    for sommet in liste_sommets:
        tmp = []
        for s in sommets_G:
            if sommet == s[0]:
                tmp.append(s)
        tmp_sorted = sorted(tmp, key=lambda element: element[1])
        sommets_tries.append(tmp_sorted)

    return sommets_tries


# initialisation des arcs du graphe G~
def arcs_G_transfome(liste_sommets, liste_arcs, sommets_G):
    arcs_G_zero = []
    arcs_G_lambda = []

    # liste des sommets (u,t) de G~ trié par sommet u
    tri_par_sommet = tri_sommets_G_transforme(liste_sommets, sommets_G)

    # ajout des arcs dont les poids sont égale à 0
    for sommets in tri_par_sommet:
        for i, s in enumerate(sommets):
            if i + 1 < len(sommets):
                arcs_G_zero.append((s, sommets[i + 1], 0))

    # ajout les arcs de poids lambda
    for sommet in liste_sommets:
        for arc in liste_arcs:
            sommet_depart = arc[0]
            sommet_arrivee = arc[1]
            date = arc[2]
            duree = arc[3]

            if sommet == sommet_depart:
                arcs_G_lambda.append(((sommet_depart, date), (sommet_arrivee, date + 1), duree))

    arcs_G = arcs_G_zero + arcs_G_lambda

    return arcs_G


# transforme le graphe G en G~
def transformation_G(liste_sommets, liste_arcs):
    # liste des sommets de G~ au format (u, t) avec t la date de départ du sommet u
    sommets_G = sommets_G_transforme(liste_sommets, liste_arcs)

    # liste des arcs de G~ au format ((u, t), (v, t), p))
    # avec (u, t) le sommet de départ, (v, t) le sommet d'arrivée et p le poids de l'arc
    arcs_G = arcs_G_transfome(liste_sommets, liste_arcs, sommets_G)

    return sommets_G, arcs_G


# transforme une liste de sommets en dictionnaire
def dict_liste_sommets(liste_sommets):
    dict_sommets = {}

    for sommet in liste_sommets:
        tmp = {sommet: ''}
        dict_sommets.update(tmp)

    return dict_sommets


# génère un graphe en format dictionnaire à partir de la liste des sommets et des arcs
def dict_liste_arcs(sommets_G, arcs_G):
    graphe = dict_liste_sommets(sommets_G)

    for sommet in sommets_G:
        tmp_dict = {}
        for arc in arcs_G:
            if sommet == arc[0]:
                tmp = {arc[1]: arc[2]}
                tmp_dict.update(tmp)
        graphe[sommet] = tmp_dict

    return graphe


# renvoie la liste contenant les sommets (u, t) selon un sommet u donnée
def get_liste_sommets_selon_sommet(sommets_G, sommet):
    liste = []
    for s in sommets_G:
        if s[0] == sommet:
            liste.append(s)
    return liste


# renvoie le chemin d'arrivée au plus tôt
def algo_type1(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee):
    liste_sommets_depart = get_liste_sommets_selon_sommet(sommets_G, sommet_depart)
    liste_sommets_arrivee = get_liste_sommets_selon_sommet(sommets_G, sommet_arrivee)
    tri_sommets_arrivee = sorted(liste_sommets_arrivee, key=lambda s: s[1])

    depart = min(liste_sommets_depart, key=lambda s: s[1])
    indice_noeud_depart = indice_noeud_dans_liste(liste_noeuds, depart)
    chemin = []

    for arrivee in tri_sommets_arrivee:
        indice_noeud_arrivee = indice_noeud_dans_liste(liste_noeuds, arrivee)
        chemin = plus_court_chemin(liste_noeuds[indice_noeud_depart], liste_noeuds[indice_noeud_arrivee])
        if chemin:
            return chemin

    return chemin


# renvoie le chemin de départ au plus tard
def algo_type2(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee):
    liste_sommets_depart = get_liste_sommets_selon_sommet(sommets_G, sommet_depart)
    liste_sommets_arrivee = get_liste_sommets_selon_sommet(sommets_G, sommet_arrivee)
    tri_sommets_depart_decroissant = sorted(liste_sommets_depart, key=lambda s: s[1], reverse=True)

    chemin = []

    for depart in tri_sommets_depart_decroissant:
        indice_noeud_depart = indice_noeud_dans_liste(liste_noeuds, depart)
        for arrivee in liste_sommets_arrivee:
            indice_noeud_arrivee = indice_noeud_dans_liste(liste_noeuds, arrivee)
            chemin = plus_court_chemin(liste_noeuds[indice_noeud_depart], liste_noeuds[indice_noeud_arrivee])
            if chemin:
                return chemin

    return chemin


# renvoie le chemin le plus rapide
def algo_type3(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee):
    liste_sommets_depart = get_liste_sommets_selon_sommet(sommets_G, sommet_depart)
    liste_sommets_arrivee = get_liste_sommets_selon_sommet(sommets_G, sommet_arrivee)

    chemin = []
    combinaisons = []

    for depart in liste_sommets_depart:
        for arrivee in liste_sommets_arrivee:
            duree = arrivee[1] - depart[1]
            combinaisons.append((depart, arrivee, duree))

    tri_combinaisons = sorted(combinaisons, key=lambda element: element[2])

    for couple in tri_combinaisons:
        depart = couple[0]
        arrivee = couple[1]
        indice_noeud_depart = indice_noeud_dans_liste(liste_noeuds, depart)
        indice_noeud_arrivee = indice_noeud_dans_liste(liste_noeuds, arrivee)
        chemin = plus_court_chemin(liste_noeuds[indice_noeud_depart], liste_noeuds[indice_noeud_arrivee])
        if chemin:
            return chemin

    return chemin


# renvoie le chemin le plus court
def algo_type4(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee):
    liste_sommets_depart = get_liste_sommets_selon_sommet(sommets_G, sommet_depart)
    liste_sommets_arrivee = get_liste_sommets_selon_sommet(sommets_G, sommet_arrivee)

    depart = min(liste_sommets_depart, key=lambda s: s[1])
    indice_noeud_depart = indice_noeud_dans_liste(liste_noeuds, depart)
    all_chemins = []

    for arrivee in liste_sommets_arrivee:
        indice_noeud_arrivee = indice_noeud_dans_liste(liste_noeuds, arrivee)
        tmp_chemin = plus_court_chemin(liste_noeuds[indice_noeud_depart], liste_noeuds[indice_noeud_arrivee])
        if tmp_chemin:
            all_chemins.append(tmp_chemin)

    dist = sys.maxsize
    chemin = []

    for c in all_chemins:
        if len(c) < dist:
            dist = len(c)
            chemin = c

    return chemin


# transforme le graphe et exécute tous 4 types d'algo
def all_algo(liste_sommets, liste_arcs, sommet_depart, sommet_arrivee):
    sommets_G, arcs_G = transformation_G(liste_sommets, liste_arcs)
    dict_arcs_G = dict_liste_arcs(sommets_G, arcs_G)
    liste_noeuds = generer_graphe(dict_arcs_G)
    # affiche_graphe(dict_arcs_G)

    print("\nType I : Chemin d'arrivée au plus tôt")
    chemin_type1 = algo_type1(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee)
    print(chemin_type1)

    print("\nType II : Chemin de départ au plus tard")
    chemin_type2 = algo_type2(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee)
    print(chemin_type2)

    print("\nType III : Chemin le plus rapide")
    chemin_type3 = algo_type3(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee)
    print(chemin_type3)

    print("\nType IV : Chemin le plus court")
    chemin_type4 = algo_type4(sommets_G, liste_noeuds, sommet_depart, sommet_arrivee)
    print(chemin_type4)


# affiche le graphe (dictionnaire)
def affiche_graphe(graphe):
    for noeud in graphe:
        print(noeud)
        for arc in graphe[noeud]:
            print("\t{} : {}".format(arc, graphe[noeud][arc]))


def lire_fichier(nom_fichier):
    with open(nom_fichier, 'r') as fichier:
        liste_sommets = []
        liste_arcs = []

        for i, ligne in enumerate(fichier):
            if i == 0:
                nb_sommets = int(ligne)
            elif i == 1:
                nb_arcs = int(ligne)
            elif 1 < i < nb_sommets+2:
                liste_sommets.append(ligne.strip('\n'))
            else:
                ligne_arc = ligne.strip('()\n')
                depart, arrivee, jour, poids = ligne_arc.split(', ')
                liste_arcs.append((depart, arrivee, int(jour), int(poids)))

    return liste_sommets, liste_arcs


if __name__ == '__main__':
    liste_sommets = ['a', 'b', 'c', 'f', 'g']  # liste des sommets de G

    # liste des arcs du graphe G au format (u, v, t, lambda)
    # avec u le sommet de départ, v le sommet d'arrivée, t la date de départ et lambda la durée du vol
    liste_arcs = [('a', 'b', 1, 1), ('a', 'b', 2, 1), ('a', 'c', 2, 1), ('a', 'c', 4, 1),
                  ('b', 'f', 5, 1),
                  ('c', 'f', 6, 1), ('c', 'g', 7, 1)]

    nb_sommets = len(liste_sommets)  # nombre de sommets de G
    nb_arcs = len(liste_arcs)  # nombre d'arcs de G

    liste_sommets_exemple, liste_arcs_exemple = lire_fichier("exemple1.txt")
    liste_sommets_exemple2, liste_arcs_exemple2 = lire_fichier("exemple3.txt")

    all_algo(liste_sommets, liste_arcs, 'a', 'f')
    print("\n--------------------")
    all_algo(liste_sommets_exemple, liste_arcs_exemple, 'a', 'k')
    print("\n--------------------")
    all_algo(liste_sommets_exemple2, liste_arcs_exemple2, 'a', 'i')


```

```python
import gurobipy as gp
import networkx as nx

def GraphFromFile(filename):

# On Lit le fichier texte et on crée le multigraphe dirigé correspondant
    G = nx.MultiDiGraph()
    
    with open(filename, encoding='utf8') as f:
        for line in f: 
            if line[0] == 's':
                line_split = line.split(" ")
                G.add_node(str(line_split[1]))
            if line[0] == 'a':
                line_split = line.split(" ")
                u, v, t, l = line_split[2][1:-2].split(",")
                G.add_edge(u, v, date=int(t), durée=int(l))
    return G

#---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def GraphFromMultiGraph(MG):
    G = nx.DiGraph()

    # Ajout de sommets
    for x, y, attribut in MG.edges(data=True):
        t = attribut["date"]
        l = attribut["durée"]
        xOut = (x, t)
        yIn = (y, t + l)
        G.add_node(xOut)
        G.add_node(yIn)
        G.add_edge(xOut, yIn, weight=l)

    # Ajout de 0 arête pondérée
    ListeSommets = list(G.nodes)
    ListeSommets = sorted(ListeSommets, key=lambda item: (item[0], item[1]))

    for i in range(len(ListeSommets)-1):
        if ListeSommets[i][0] == ListeSommets[i+1][0]:
            G.add_edge(ListeSommets[i], (ListeSommets[i+1]), weight=0)
            
    return G

#---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def CheminLePlusCourtPL(G, depart, arrivée, A=0, B=float("Inf")):

    with gp.Env(empty=True) as env:

        # Supprimer les output log de gurobi
        env.setParam('OutputFlag', 0)
        env.start()
        with gp.Model(env=env) as m:
            l_arretes = list(G.edges(data=True))
            nb_arretes = len(l_arretes)
            
            # variables de décision
            x = []
            for i in range(len(l_arretes)):
                x.append(m.addVar(vtype=gp.GRB.BINARY, lb=0, name=f"x{i}"))
            
            # mis à jour du modèle
            m.update()

            # définir la fonction objectif
            obj_func = gp.LinExpr()
            obj_func = 0
            for i in range(len(l_arretes)):
                obj_func += l_arretes[i][2]["weight"] * x[i]

            m.setObjective(obj_func, gp.GRB.MINIMIZE)

            # définir les contraintes
            for node in G.nodes():

                # Nombre d'arêtes d'entrée = nombre d'arêtes de sortie 
                if node[0] != depart and node[0] != arrivée:
                    index_arrete_in = [i for i in range(nb_arretes) if l_arretes[i][1] == node]
                    index_arrete_out = [i for i in range(nb_arretes) if l_arretes[i][0] == node]
                    m.addConstr(gp.quicksum(x[i] for i in index_arrete_in) - gp.quicksum(x[i] for i in index_arrete_out) == 0, f"Contrainte noeud {node}")

                # Aucun bord n'entre dans le nœud de départ, aucun bord ne sort du nœud de départ avant A.
                elif node[0] == depart:
                    index_arrete_in = [i for i in range(nb_arretes) if l_arretes[i][1] == node]
                    index_arrete_avant_A_out = [i for i in range(nb_arretes) if l_arretes[i][0] == node and node[1] < A]
                    edge_index = index_arrete_in + index_arrete_avant_A_out
                    if edge_index:
                        m.addConstrs((x[i] == 0 for i in index_arrete_in + index_arrete_avant_A_out), f"Contrainte noeud {node}")

                # Aucun bord ne sort du noeud d'arrivée, aucun bord n'entre dans le noeud d'arrivée après B.
                elif node[0] == arrivée:
                    index_arrete_out = [i for i in range(nb_arretes) if l_arretes[i][0] == node]
                    index_arrete_après_B_in = [i for i in range(nb_arretes) if l_arretes[i][1] == node and node[1] > B]
                    edge_index = index_arrete_out + index_arrete_après_B_in
                    if edge_index:
                        m.addConstrs((x[i] == 0 for i in edge_index), f"Contrainte noeud {node}")

            # Pas d'arête entre le nœud de départ
            index_arrete_entre_noeud_depart = [i for i in range(nb_arretes) if l_arretes[i][0][0] == depart and l_arretes[i][1][0] == depart]
            if index_arrete_entre_noeud_depart:
                m.addConstrs((x[i] == 0 for i in index_arrete_entre_noeud_depart), f"Contrainte noeud de départ")

            # Pas d'arête entre le noeud d'arrivée
            index_arrete_entre_noeud_arrivee = [i for i in range(nb_arretes) if l_arretes[i][0][0] == arrivée and l_arretes[i][1][0] == arrivée]
            if index_arrete_entre_noeud_arrivee:
                m.addConstrs((x[i] == 0 for i in index_arrete_entre_noeud_arrivee), f"Contrainte noeud d'arrivé")

            # Il y a un bord qui sort de l'ensemble du noeud de départ
            out_departure_node_index = [i for i in range(nb_arretes) if l_arretes[i][0][0] == depart and l_arretes[i][0][1] > A]
            m.addConstr(gp.quicksum(x[i] for i in out_departure_node_index) == 1, f"Contrainte sur l'arrête qui sort du nœud de départ")

            # Il y a une arête qui entre dans l'ensemble du noeud d'arrivée
            in_arrival_node_index = [i for i in range(nb_arretes) if l_arretes[i][1][0] == arrivée and l_arretes[i][1][1] < B]
            m.addConstr(gp.quicksum(x[i] for i in in_arrival_node_index) == 1, f"Contrainte sur l'arrête qui entre dans le nœud d'arrivé")

            # Resolution
            m.optimize()

            path = []
            for i in range(nb_arretes):
                try:
                    if x[i].x == 1:
                        path.append(l_arretes[i])
                except AttributeError:
                    return None
            
            return path

#G2 = GraphFromMultiGraph(GraphFromFile("graphe.txt"))
#path =  CheminLePlusCourtPL(G2, "a", "l")
#print(f"Plus court chemin de a vers l : {path}")
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np

def MultiGraph(nb_sommet, nb_arrete, durée):

    # générer un multigraphe dirigé en fonction d'un nombre de sommets et la probabilité d'apparition d'une arête et une durée fixe
    
    G = nx.MultiDiGraph()
    
    # Sommets
    for i in range(1, nb_sommet+1):
        G.add_node(str(i))
    if len(G.nodes()) == 1:
        return G

    # # On diminu e tant qu'on a pas ajouté toutes les arrêtes du graphe 
    startDate = 0
    while nb_arrete > 0: 
        for i in range(1, nb_sommet+1):
            depDate = startDate+1
            for j in range(i+1, i+3):
                if nb_arrete <= 0: 
                    return G
                if str(j) not in G.nodes():
                    break
                depDate+=1
                dateArr = depDate + j 
                G.add_edge(str(i), str(j), date = dateArr, durée=durée)
                nb_arrete -= 1
                
        startDate += 1
    return G

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def TriSommet(G):

    sommet = dict()
    for v,t in G.nodes():
        if v not in sommet:
            sommet[v] = [(v,t)]
        else:
            sommet[v].append((v,t))
    for v in sommet:
        sommet[v].sort(key=lambda x:x[1])
    return sommet

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def StressTestNB_Sommets():

    durée=1
    occurences=50
    max_nb_sommet=100
    minDate=0
    maxDate=float("Inf")

    TempsResultant = np.zeros(max_nb_sommet)
    nb_sommet = np.zeros(max_nb_sommet)

    for j in range(occurences):
        for i in range(1, max_nb_sommet+1, 1):
            nb_sommet[i-1] = i

            # On génère G
            Gtest = GraphFromMultiGraph(MultiGraph(i, 2*i, durée)) 

            t0 = time.time()
            SommetsTriés = TriSommet(Gtest)
            TrierLeTemp = time.time() - t0

            if SommetsTriés == {}:
                continue

            t = time.time()
            CheminLePlusCourtPL(Gtest, SommetsTriés['1'][0][0], SommetsTriés[str(i-1)][0][0], minDate, maxDate)
            TempsResultant[i-1] += time.time() - t + TrierLeTemp
    
    return (TempsResultant/occurences), nb_sommet

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def StressTestNB_Arcs():

    occurences=100
    nb_sommet=50
    durée=1
    minDate=0
    maxDate=float("Inf")

    # Nous avons besoin de 2n arêtes pour nous assurer que le dernier sommet est atteint.
    maxE = 6 * nb_sommet
    minE = 2 * nb_sommet
    taille = maxE - minE + 1
    
    TempsResultant = np.zeros(taille)
    nb_arcs = np.zeros(taille)

    for j in range(occurences):
        for i in range(minE,maxE+1,1):
            nb_arcs[i-minE] = i

            # On génère G
            Gtest = GraphFromMultiGraph(MultiGraph(nb_sommet, i, durée)) 
            nodes = list(MultiGraph(nb_sommet, i, durée).nodes())

            # Algorithmes
            t0 = time.time()
            SommetsTriés = TriSommet(Gtest)

            if SommetsTriés == {}:
                continue

            t = time.time()
            CheminLePlusCourtPL(Gtest, SommetsTriés['1'][0][0], nodes[nb_sommet - 1], minDate, maxDate)
            TempsResultant[i-minE] += time.time() - t

    return (TempsResultant/occurences), nb_arcs

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
   
def StressTest_Durée():

    occurences=100
    nb_sommet=20
    max_durée=20
    minDate=0
    maxDate=float("Inf")
    
    # Nous avons besoin de 2 x nb_sommet pour nous assurer que le dernier sommet est accessible.
    nb_edges = 2 * nb_sommet

    TempsResultant = np.zeros(max_durée + 1)
    durée = np.zeros(max_durée + 1)

    for j in range(occurences):
        for i in range(5, max_durée + 6, 1):
            durée[i-5] = i

            # On génère G 
            Gtest = GraphFromMultiGraph(MultiGraph(nb_sommet, nb_edges, i)) 

            t0 = time.time()
            SommetsTriés = TriSommet(Gtest)
            TrierLeTemps = time.time() - t0

            if SommetsTriés == {}:
                continue

            t = time.time()
            CheminLePlusCourtPL(Gtest, SommetsTriés['1'][0][0], SommetsTriés[str(nb_sommet - 1)][0][0], minDate, maxDate)
            TempsResultant[i-5] += time.time() - t + TrierLeTemps

    return (TempsResultant/occurences), durée





TempsResultant1, NombreDeSommets = StressTestNB_Sommets()
TempsResultant2, NombreArcs = StressTestNB_Arcs()
TempsResultant3, Duree = StressTest_Durée()

def plot_test(x, y, xtitle, ytitle="Temps d'execution",title="Test du PL"):
    
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    plt.legend()
    plt.show()

plot_test(TempsResultant1, NombreDeSommets,"Nb_sommet")


```
